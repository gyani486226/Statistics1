# covvariance

```python
df.cov(numeric_only = True)
# covariance rangw can be anything
#between fare and passengers the covariance value is 0.106199 it means they are positive correlated
#between tolls and passengers the covariance is -0.004946 it means the are negative correleated (means tolls increase, passengers decreases)
```

# correlation 

```python
# the range of correleation is (-1 t0 1), dimensionless quantity
df.corr(numeric_only = True)

# the correlation between distance and tip  is 0.45 i.e. if tip increases 45% chance distance is also increases
```

```python
corr= df.corr(numeric_only = True)
sns.heatmap(corr,  annot = True)
```
